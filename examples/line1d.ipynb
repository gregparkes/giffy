# Line1D

- Requires X (1d or 2d), Y (2d)
- By default the rows of Y is the time axis if not specified.
- Can use linear interpolation if you need more or less timesteps.
- Moves the line incrementally.

```python
x2 = np.linspace(-3, 3, 91)
t = np.linspace(0, 25, 30)
X2, T2 = np.meshgrid(x2, t)
sinT2 = np.sin(2*np.pi*T2/T2.max())
F = 0.9*sinT2*np.sinc(X2*(1 + sinT2))
#df1_5 = pd.DataFrame(np.vstack((x2.reshape(1,91), F)))
```

#### Basic Plot

```python
gly.Line(Y=F).plot()
```

#### Interpolation

```python
gly.Line(Y=F, X=x2).mark(x='time', y='force', ls='--').lerp(2).plot()
```

```python
gly.Line(Y=F, X=x2).mark(x='time', y='force', marker='.', ls='').plot()
```

### Alpha

```python
gly.Line(Y=F, X=x2).mark(x='time', y='force', a=.2).plot()
```
